# II. Feature Engineering

## Generate weekday and time period by hour

```python
# check the time range
print(np.max(df.hour))
print(np.min(df.hour))
```

```python
# define a function to get the weekday of the time
def get_date(hour):
    y = '20'+str(hour)[:2]
    m = str(hour)[2:4]
    d = str(hour)[4:6]
    return y + '-' + m + '-' + d
```

```python
# get the weekday
df['weekday'] = pd.to_datetime(df.hour.apply(get_date)).dt.day_name()
df['weekday']
```

```python
# define a function to get the hour of the time
def get_hour(hour):
    h = str(hour)[6:8]
    h = int(h)
    return h
```

```python
# get the hour
df['hour'] = df.hour.apply(get_hour)
df['hour']
```

```python
# transfer to time period
df.loc[df.hour.between(6,12), 'time_period'] = 'morning'
df.loc[df.hour.between(13,18), 'time_period'] = 'afternoon'
df.loc[df.hour.between(19,24), 'time_period'] = 'night'
df.loc[df.hour.between(0,5), 'time_period'] = 'mid_night'
```

```python
# drop columns won't be used for modeling
df = df.drop(['id','hour'], axis=1)
```

## Transfer categories with only one unique values to others

```python
good_col = ['click','C1','banner_pos','device_type','device_conn_type','C15','C16','C18','weekday','time_period']
```

```python
for i in [x for x in df.columns if x not in good_col]:
    cnt = df[i].value_counts()
    row = cnt[cnt == 1].index.tolist()
    df.loc[df[i].isin(row),i] = 'others'
```
